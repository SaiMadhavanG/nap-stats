```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import pandas as pd
```

Layerwise bar plots

```python
def get_layerwise_plot(naps_path, label, n_layers):
    naps = json.load(open(naps_path))
    df = pd.DataFrame(columns=['layer', 'count', 'if_activated'])
    nap = naps[str(label)]
    for layer in range(n_layers):
        cnt_A = sum([1 if idx[0]==layer else 0 for idx in nap['A']['indices']])
        cnt_D = sum([1 if idx[0]==layer else 0 for idx in nap['D']['indices']])
        df.loc[len(df)] = [layer, cnt_A, 'A']
        df.loc[len(df)] = [layer, cnt_D, 'D']
    assert df['count'].sum() == nap['A']['len'] + nap['D']['len']
    sns.barplot(df, x = 'layer', y = 'count', hue='if_activated')
    delta = naps['config']['delta']
    net = naps['config']['net']
    plt.title(f"Layer wise stats for Î´ = {delta}; net = {net}; label = {label}")
    plt.show()
    return df
```

```python
for i in range(10):
    df = get_layerwise_plot("./NAPs/256x4_delta1.json", i, 4)
```
