```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
import pandas as pd
```

Labelwise bar plots for A/D

```python
def labelwise_barplot(naps_path):
    naps = json.load(open(naps_path))
    df = pd.DataFrame(columns=['label', 'count', 'if_activated'])
    for label in range(10):
        cnt_A = naps[str(label)]['A']['len']
        cnt_D = naps[str(label)]['D']['len']
        df.loc[len(df)] = [label, cnt_A, 'A']
        df.loc[len(df)] = [label, cnt_D, 'D']
    sns.barplot(df, x = 'label', y = 'count', hue='if_activated')
    delta = naps['config']['delta']
    net = naps['config']['net']
    plt.title(f"Label wise stats for Î´ = {delta}; net = {net}")
    plt.show()
    return df
```

```python
df = labelwise_barplot("./NAPs/256x4_delta1.json")
```
